```python
import json

prompts = [1, 3, 7, 9, 10] #With 117 or 118 filtered
manual_prompts = [1, 2, 3, 4]

prompt_files = []

for i in manual_prompts:
    path = f"/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics/gpt/4o-mini_manual_prompt_{i}_75_mini_actionswap/filtered_prompts.json"
    prompt_files.append(path)

common_videos = None  # start with None, so first file initializes it

for i in prompt_files:
    with open(i, "r") as f:
        data = json.load(f)

    # collect all paths from this file
    videos = {item["action_swap_path"] for item in data}


    # first time: initialize
    if common_videos is None:
        common_videos = videos
    else:
        # keep only those present in both sets
        common_videos = common_videos.intersection(videos)

print("Number in common:", len(common_videos))
print("Common video paths:", list(common_videos)[:10])  # preview first 10
```

```python
import json
with open("/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics/filtered_119.json", "r") as f:
    filtered_119 = json.load(f)


mapping = {"A":1, "B":2, "C":3, "D":4, "E":5}

for i in range(1, 5):
    print("=======PROMPT ", i)
    path = f"/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics/gpt/4o-mini_manual_prompt_{i}_75_mini_actionswap/results.json"
    filtered_prompts = f"/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics/gpt/4o-mini_manual_prompt_{i}_75_mini_actionswap/filtered_prompts_results.json"

    pred_human_azure = 0
    pred_bg_azure = 0
    pred_human_real = 0
    pred_bg_real = 0
    total = 0

    with open(path, "r") as f: #results for 
        data = json.load(f)

    with open(filtered_prompts, "r") as f: #results for 
        data_filtered = json.load(f)

    for key, value in data.items():
        if value["flagged"]:
            continue
        if value["action_swap_path"] in filtered_119:
            # print("here0")
            total += 1
            print(value)
            print(value["prompt"], value["response"])
            if value["human_choice"] == mapping[value['response'][0]]:
                # print("here1")
                pred_human_azure += 1
            if value["background_choice"] == mapping[value["response"][0]]:
                # print("here2")
                pred_bg_azure += 1

    for key, value in data_filtered.items():
        if value["action_swap_path"] in filtered_119:
            # print("here3")
            if value["human_choice"] == value["llm_choice"]:
                # print("here4")
                pred_human_real += 1
            elif value["background_choice"] == value["llm_choice"]:
                # print("here5")
                pred_bg_real += 1

    print("pred human azure/real, bg azure/real, total:", pred_human_azure, pred_human_real, pred_bg_azure, pred_bg_real, total)




```
