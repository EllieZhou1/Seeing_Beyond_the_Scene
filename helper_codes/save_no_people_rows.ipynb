```python
import pandas as pd

# Load both CSVs
df_full = pd.read_csv("dataset/HAT_minikinetics50_validation.csv")
df_subset = pd.read_csv("dataset/segmented_minikinetics50/segmented_minikinetics50_validation.csv")

# Choose the columns that uniquely identify a row (e.g., 'youtube_id', 'time_start', 'time_end')
key_cols = ['youtube_id']

# Find rows in df_full that are NOT in df_subset
mask = ~df_full[key_cols].apply(tuple, 1).isin(df_subset[key_cols].apply(tuple, 1))
df_missing = df_full[mask]

# Save to a new CSV
df_missing.to_csv("dataset/segmented_minikinetics50/no_human_rows_in_minikinetics50_validation.csv", index=False)
print(f"Saved {len(df_missing)} missing rows to missing_rows.csv")
```
