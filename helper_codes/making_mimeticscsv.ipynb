```python
import os
import csv
```

```python
mimetics_dir = "/n/fs/visualai-scr/Data/Mimetics2"
output_csv = "dataset/mimetics.csv"
```

```python
rows = []
```

```python
for label in os.listdir(mimetics_dir):
    label_path = os.path.join(mimetics_dir, label)
    print("label path", label_path)

    #skip if not a real path
    if not os.path.isdir(label_path):
        continue

    for filename in os.listdir(label_path):
        full_path = os.path.join(label_path, filename)
        print("ful path", full_path)
        num_files = sum(1 for entry in os.scandir(full_path) if entry.is_file())
        
        folder_name = os.path.basename(full_path)  # e.g., 'EOQ0ST_d6_mQ_000074_000084'
        parts = folder_name.rsplit('_', 2)        # splits from the right, max 2 splits
        youtube_id = parts[0]                     # everything before the last two underscores
        print("parts", parts)
        time_start = int(parts[1])
        time_end = int(parts[2])

        rows.append({
            "label": label,
            "youtube_id": youtube_id,
            "full_path": full_path,
            "num_files": num_files,
            "time_start": time_start,
            "time_end": time_end
        })
```

```python
os.chdir("/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics")
```

```python
with open(output_csv, "w", newline='') as csvfile:
    fieldnames = ["label", "youtube_id", "full_path", "num_files", "time_start", "time_end"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in rows:
        writer.writerow(row)

print(f"Wrote {len(rows)} rows to {output_csv}")
```
