```python
import pandas as pd

# Set file paths
input_path = "/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics/dataset/minikinetics50/minikinetics50_train_all.csv.bak"
val_output_path = "/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics/dataset/minikinetics50/minikinetics50_validation_all.csv"
train_output_path = "/n/fs/visualai-scr/temp_LLP/ellie/slowfast_kinetics/dataset/minikinetics50/minikinetics50_train_all.csv"

# Load the data
df = pd.read_csv(input_path)

# Shuffle and split
val_df = df.sample(frac=0.2, random_state=42)
train_df = df.drop(val_df.index)

# Save the splits
val_df.to_csv(val_output_path, index=False)
train_df.to_csv(train_output_path, index=False)

print(f"Validation set saved to {val_output_path} with {len(val_df)} rows.")
print(f"Training set saved to {train_output_path} with {len(train_df)} rows.")
```
