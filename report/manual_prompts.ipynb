```python
import matplotlib.pyplot as plt

# Data
labels = [
    "Prompt 1",
    "Prompt 2",
    "Prompt 3",
    "Prompt 4"
]
human_accuracy = [39.53862661, 34.54935622, 41.20171674, 35.99785408]
background_error = [50.8583691, 46.88841202, 46.72746781, 52.7360515]
colors = ['cornflowerblue', 'salmon', 'lightseagreen', 'red']

# Create scatter plot
plt.figure(figsize=(7, 7), dpi=300)

plt.minorticks_on()
plt.grid(which='minor', linestyle=":", zorder=0)
plt.grid(True)
plt.xticks([30, 35, 40, 45, 50, 55])
plt.yticks([30, 35, 40, 45, 50, 55])

for i in range(4):
    plt.scatter(human_accuracy[i], background_error[i], s=250, color=colors[i], edgecolor='black', label=labels[i], zorder=3)

# Draw arrows from Prompt 1 (blue dot) to all others
x0, y0 = human_accuracy[0], background_error[0]
for i in range(1, 4):
    plt.annotate(
        "", 
        xy=(human_accuracy[i], background_error[i]),   # target point
        xytext=(human_accuracy[0], background_error[0]),  # blue dot origin
        arrowprops=dict(
            arrowstyle="->",
            lw=2,
            color="gray",
            shrinkA=15,   # shrink distance from arrow tail (blue dot)
            shrinkB=15    # shrink distance from arrow head (target dot)
        )
    )

# Add annotations (emoji-like)

# Axis labels and title
plt.xlabel("SHacc (%)", fontsize=15)
plt.ylabel("SBerr (%)", fontsize=15)
plt.title("GPT 4o-mini Manually Crafted Prompt Performance: SBerr vs. SHacc", fontsize=15)

plt.legend(labelspacing=1.2)
plt.xlim(30, 55)
plt.ylim(30, 55)
plt.savefig("manual_prompt.png", dpi=300, bbox_inches='tight')
plt.show()
```
