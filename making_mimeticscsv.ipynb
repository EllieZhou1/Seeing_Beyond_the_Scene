```python
import os
import csv
```

```python
mimetics_dir = "/n/fs/visualai-scr/Data/Mimetics/Mimetics_videos"
output_csv = "dataset/mimetics.csv"
```

```python
rows = []
```

```python
for label in os.listdir(mimetics_dir):
    label_path = os.path.join(mimetics_dir, label)

    #skip if not a real path
    if not os.path.isdir(label_path):
        continue

    for filename in os.listdir(label_path):
        full_path = os.path.join(label_path, filename)

        rows.append({
            "label": label,
            "full_path": full_path
        })
```

```python
with open(output_csv, "w", newline='') as csvfile:
    fieldnames = ["label", "full_path"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in rows:
        writer.writerow(row)

print(f"Wrote {len(rows)} rows to {output_csv}")
```
