```python
import matplotlib.pyplot as plt

# Data
labels = [
    "Prompt 1\n",
    "Prompt 2\n",
    "Prompt 3\n"
]
human_accuracy = [0.3640256959, 0.378947, 0.402542]
background_error = [0.4668094218, 0.509474, 0.457627]

# Create scatter plot
plt.figure(figsize=(7, 7))
plt.scatter(human_accuracy[0], background_error[0], s=200, color='cornflowerblue', label='Prompt 1')
plt.scatter(human_accuracy[1], background_error[1], s=200, color='salmon', label='Prompt 2')
plt.scatter(human_accuracy[2], background_error[2], s=200, color='lightseagreen', label='Prompt 3')

# Axis labels and title
plt.xlabel("Human Accuracy", fontsize=15)
plt.xticks([0.35, 0.40, 0.45, 0.50, 0.55])
plt.yticks([0.35, 0.40, 0.45, 0.50, 0.55])
plt.minorticks_on()
plt.grid(which='minor', linestyle=":")
plt.ylabel("Background Error", fontsize=15)
plt.title("GPT 4o-mini Prompt Performance: Background Error vs. Human Accuracy", fontsize=15)
plt.grid(True)
plt.legend(labelspacing=1.2)
plt.xlim(0.35, 0.55)
plt.ylim(0.35, 0.55)

plt.tight_layout()
plt.show()
```
