```python
#Difference between original and mixed data
#On Kinetics


import matplotlib.pyplot as plt
import numpy as np

# Data
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Slow-Only', 'Segmented', 'Sum', 'Stack', 'Weighted Focus']

metric_orig = np.array([0.4993963783, 0.2346076459, 0.5215291751, 0.5150905433, 0.5203219316])
metric_mixed = np.array([0.4655935614, 0.2325955734, 0.3971830986, 0.429778672, 0.2386317907])
x = np.arange(len(labels))
bar_width = 0.6

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Draw bars with correct stacking order and colors
for i in range(len(x)):
    a = metric_orig[i]
    b = metric_mixed[i]
    bottom = min(a, b)
    height = abs(a - b)

    # Base bar: whichever is smaller
    if a < b:
        ax.bar(x[i], a, width=bar_width, color='lightsteelblue', label='Original' if i == 0 else "")
        ax.bar(x[i], height, bottom=a, width=bar_width, color='royalblue', label='Augmented Data' if i == 0 else "")
    else:
        ax.bar(x[i], b, width=bar_width, color='royalblue', label='Augmented Data' if i == 0 else "")
        ax.bar(x[i], height, bottom=b, width=bar_width, color='lightsteelblue', label='Original' if i == 0 else "")

    # Add arrow from A to B
    ax.annotate(
        '', 
        xy=(x[i], b), 
        xytext=(x[i], a), 
        arrowprops=dict(arrowstyle='->', color='black', lw=1.5),
        ha='center'
    )

# Format plot

ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=15)
ax.set_ylim(0.0, 0.6)
ax.set_ylabel("Accuracy", fontsize=15)
ax.set_title("Performance on Mini-Kinetics 50 with Augmented Training Data", fontsize=15)
ax.legend()
plt.tight_layout()
plt.show()
```
