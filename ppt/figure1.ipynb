```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = [
    "skiing (not slalom or crosscountry)",
    "canoeing or kayaking",
    "shooting goal (soccer)",
    "tying tie",
    "punching person (boxing)",
    "eating ice cream"
]

bg_top1_rate = [0.455, 0.436, 0.407, 0.185, 0.108, 0.098]
human_top1_rate = [0.0, 0.0, 0.037, 0.685, 0.432, 0.275]

# X locations for groups
x = np.arange(len(labels))
width = 0.35  # width of the bars

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, human_top1_rate, width, label='Human Accuracy', color='cornflowerblue')
bars2 = ax.bar(x + width/2, bg_top1_rate, width, label='Background Error', color='lightsalmon')

# Add vertical dotted line between third and fourth group
separator_x = (x[2] + x[3]) / 2
ax.axvline(x=separator_x, linestyle='dotted', color='gray', linewidth=2)

# Labels and formatting
ax.set_ylabel('Accuracy')
ax.set_title('Human Accuracy and Background Error on HAT Action Swap, grouped by Human Action Label')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=20, ha='right')
ax.legend()

# Annotate bars with values
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.3f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
