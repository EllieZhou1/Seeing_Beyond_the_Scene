```python
#+ Sum-Concat and Stack-Concat
# On Kinetics

import matplotlib.pyplot as plt
import numpy as np

x = [0, 0.15, 0.30, 0.45]
labels = ['Slow-Only', 'Segmented', 'Sum-Concat', 'Stack-Concat']
y = [0.4993963783, 0.2346076459, 0.5215291751, 0.5150905433]

plt.figure(figsize=(8, 5))
plt.ylim(0.0, 0.55)
plt.xticks(x, labels, fontsize=15)
bars = plt.bar(x, y, label="Human Accuracy", width=0.1, color='cornflowerblue')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f'{height:.4f}', ha='center')


plt.ylabel("Accuracy", fontsize=15)
plt.title("Performance on Kinetics", fontsize=15)

plt.show()
```
