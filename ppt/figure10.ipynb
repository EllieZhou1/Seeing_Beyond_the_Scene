```python
#Segmented
#Kinetics


import matplotlib.pyplot as plt
import numpy as np

x = [0, 0.15, 0.30]
labels = ['Slow-Only', 'Segmented', ' ']
y = [0.4993963783, 0.2346076459, 0]

plt.figure(figsize=(8, 5))
plt.ylim(0.0, 0.55)
plt.xticks(x, labels, fontsize=15)
plt.bar(x, y, label="Human Accuracy", width=0.1, color='cornflowerblue')
plt.ylabel("Accuracy", fontsize=15)
plt.title("Performance on Kinetics", fontsize=15)

plt.show()
```
