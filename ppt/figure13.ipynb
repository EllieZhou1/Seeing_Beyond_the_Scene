```python
# + Sum-Concat and Stack-Concat
# HAT Action Swap


import matplotlib.pyplot as plt
import numpy as np

x = np.array([0.15, 0.35, 0.55, 0.75])
print(x)
labels = ['Slow-Only', 'Segmented', 'Sum-Concat', 'Stack-Concat']
y_human = [0.09617706237, 0.2334004024, 0.1275653924, 0.1279678068]
y_bg = [0.2342052314, 0.02092555332, 0.2036217304, 0.1979879276]

plt.figure(figsize=(8, 5))
plt.ylim(0.0, 0.3)
plt.xticks(x, labels, fontsize=15)

width=0.08
plt.bar(x-width/2, y_human, label="Human Accuracy", width=width, color='cornflowerblue')
plt.bar(x+width/2, y_bg, label="Background Error", width=width, color='lightsalmon')

plt.ylabel("Accuracy", fontsize=15)
plt.title("Performance on HAT Action-Swap", fontsize=15)
plt.legend()

plt.show()
```
