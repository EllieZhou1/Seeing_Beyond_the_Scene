```python
import matplotlib.pyplot as plt
import numpy as np

# Data
prompts = ['Prompt #1\n“What is the action being performed?”',
           'Prompt #2\n“a photo of a human ...”',
           'Prompt #3\nHuman-Focused']
one_frame = [0.475456, 0.447405, 0.475456]
two_frame = [0.503506, 0.475456, 0.499299]

x = np.arange(len(prompts))  # the label locations
width = 0.35  # width of the bars

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(x - width/2, one_frame, width, label='1 Frame', color='cornflowerblue')
bar2 = ax.bar(x + width/2, two_frame, width, label='2 Frames', color='lightsalmon')

# Labels and titles
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy by Prompt Type and Frame Count')
ax.set_xticks(x)
ax.set_xticklabels(prompts)
ax.legend()

# Annotate bar values
for bar in bar1 + bar2:
    height = bar.get_height()
    ax.annotate(f'{height:.4f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
