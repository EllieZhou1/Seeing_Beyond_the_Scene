```python
#Baseline Slow-Only on Mini-Kinetics and HAT (full, not mcq)
#Mini-Kinetics: 0.4993963783
# HAT: 0.09617706237 (human), 0.2342052314 (bg)

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2)
labels = ['Slow-Only', ' ']
y_human = [0.09617706237, 0]
y_bg = [0.2342052314, 0]


width=0.35
plt.figure(figsize=(4, 5))
plt.bar(x-width/2, y_human, width=width, color='cornflowerblue', label="Human Accuracy")
plt.bar(x+width/2, y_bg, width=width, color='lightsalmon', label="Background Error")


plt.xticks(x, labels)
plt.ylim(0.0, 0.55)
plt.ylabel("Accuracy")
plt.title("Performance on HAT Action Swap")
plt.legend()

plt.show()
```
