```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Slow-Only', 'CLIP', 'SIGLIP2', 'Intern-VL3', 'GPT 4o-mini']
human = [0.2869215292, 0.416901, 0.2668, 0.403622, 0.3905047049] #For Intern-VL3, I chose the 8B-8 frame version (bc slow-only and 4o-mini are all w/ 8 frame)
#For gpt, this is the version where we test on everything (all 2485)

background = [0.4736418511, 0.5087, 0.5827, 0.504225, 0.5115483319]

x = np.arange(len(labels))  # the label locations
width = 0.35  # width of the bars

# Plot bars side by side
plt.figure(figsize=(6, 5))
plt.bar(x - width/2, human, width, label='Human Accuracy', color='cornflowerblue')
plt.bar(x + width/2, background, width, label='Background Error', color='lightsalmon')

# Labels and legend
plt.xticks(x, labels)
plt.ylabel("Accuracy")
plt.title("Performance on HAT Action-Swap")
plt.legend()
plt.tight_layout()
plt.show()
```
