```python
#+Weighted Focus
#On Kinetics
import matplotlib.pyplot as plt
import numpy as np

x = [0, 0.20, 0.40, 0.60, 0.80]
labels = ['Slow-Only', 'Segmented', 'Sum', 'Stack', 'Weighted Focus']
y = [0.4993963783, 0.2346076459, 0.5215291751, 0.5150905433, 0.5203219316]

plt.figure(figsize=(11, 5))
plt.ylim(0.0, 0.55)
plt.xticks(x, labels, fontsize=15)
bars = plt.bar(x, y, label="Human Accuracy", width=0.13, color='cornflowerblue')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f'{height:.4f}', ha='center')


plt.ylabel("Accuracy", fontsize=14)
plt.title("Performance on Kinetics", fontsize=15)

plt.show()
```
