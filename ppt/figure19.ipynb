```python
import matplotlib.pyplot as plt
import matplotlib.transforms as mtransforms
import numpy as np

# Data

x = np.arange(0, 2.0, 0.2)
labels = ['Slow-Only', 'Segmented', 'Sum-Concat', 'Stack-Concat', 'Weighted Focus',
          'Slow-Only: Aug', 'Segmented: Aug', 'Sum-Concat: Aug', 'Stack-Concat: Aug', 'Weighted Focus: Aug']
y_human = [0.09617706237, 0.2334004024, 0.1275653924, 0.1279678068, 0.1279678068, 0.1472837022, 0.2189134809, 0.2426559356, 0.2317907445, 0.2253521127]

y_bg = [0.2342052314, 0.02092555332, 0.2036217304, 0.1979879276, 0.1963782696, 0.1440643863, 0.02173038229, 0.07645875252, 0.08571428571, 0.0185110664]

# Create figure and axis
fig, ax = plt.subplots(figsize=(11, 5))
ax.set_ylim(0.0, 0.3)

# Bar plot

width=0.08
bars1 = ax.bar(x-width/2, y_human, label="Human Accuracy", width=width, color='cornflowerblue')
bars2 = ax.bar(x+width/2, y_bg, label="Background Error", width=width, color='lightsalmon')

# Add text labels above bars
for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f'{height:.4f}', ha='center', fontsize=8)

for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f'{height:.4f}', ha='center', fontsize=8)

# Set x-ticks and labels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12, rotation=30)

# Shift labels to the left using ScaledTranslation
offset_pts = -15  # negative = shift left
for label in ax.get_xticklabels():
    label.set_ha('center')
    label.set_transform(label.get_transform() + mtransforms.ScaledTranslation(offset_pts / 72, 0, fig.dpi_scale_trans))

# Labels and title
ax.set_ylabel("Accuracy", fontsize=14)
ax.set_title("Performance on HAT Action-Swap", fontsize=15)
ax.legend()

plt.tight_layout()
plt.show()
```
