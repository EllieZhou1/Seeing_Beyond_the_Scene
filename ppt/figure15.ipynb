```python
# + Weighted Focus
# HAT Action Swap


import matplotlib.pyplot as plt
import numpy as np

x = np.array([0.15, 0.35, 0.55, 0.75, 0.95])
print(x)
labels = ['Slow-Only', 'Segmented', 'Sum', 'Stack', 'Weighted Focus']
y_human = [0.09617706237, 0.2334004024, 0.1275653924, 0.1279678068, 0.1279678068]
y_bg = [0.2342052314, 0.02092555332, 0.2036217304, 0.1979879276, 0.1963782696]

plt.figure(figsize=(11, 5))
plt.ylim(0.0, 0.3)
plt.xticks(x, labels, fontsize=15)

width=0.08
bars1 = plt.bar(x-width/2, y_human, label="Human Accuracy", width=width, color='cornflowerblue')
bars2 = plt.bar(x+width/2, y_bg, label="Background Error", width=width, color='lightsalmon')

for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f'{height:.4f}', ha='center')

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, f'{height:.4f}', ha='center')


plt.ylabel("Accuracy", fontsize=15)
plt.title("Performance on HAT Action-Swap", fontsize=15)
plt.legend()

plt.show()
```
